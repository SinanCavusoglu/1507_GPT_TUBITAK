```python
import os 
import pandas as pd
domain_list = os.listdir("dataset_csv")
domain_dict = {}
for domain in domain_list:
    for csv in os.listdir(f"dataset_csv/{domain}"):
        domain_dict[domain] = pd.read_csv(f"dataset_csv/{domain}/{csv}")

```

```python
df = pd.concat(domain_dict.values())
```

```python
df_dup = df.drop_duplicates(["Article_Name", "Article_Journal", "Article_Volume", "Article_PDF_URL"])
df_dup =df_dup.reset_index(drop=True)
```

```python
list_20 = df_dup.Article_Journal.value_counts()[df_dup.Article_Journal.value_counts() >= 20].index
```

```python
df_dup_20 = df_dup[df_dup.Article_Journal.isin(list_20)]
```

```python
df_dup_20=df_dup_20[df_dup_20.Searching != "İşsizlik"]
```
